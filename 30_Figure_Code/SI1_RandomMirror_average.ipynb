```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import math
from matplotlib.ticker import *

# load Data
ExcelFile= '/Users/hata_s/Dropbox/01_Research/02_MF_TimeSeries/03_Matlab/02_Result/SI1_Data.xlsx'
DF_RP = pd.read_excel(ExcelFile, sheet_name='RP',skiprows=1).set_index('No')
DF_RP_Value = pd.read_excel(ExcelFile, sheet_name='RP_Value',skiprows=1).set_index('No')
DF_FD = pd.read_excel(ExcelFile, sheet_name='FD',skiprows=1).set_index('No')
DF_FD_Value = pd.read_excel(ExcelFile, sheet_name='FD_Value',skiprows=1).set_index('No')
DF_CUin = pd.read_excel(ExcelFile, sheet_name='CUin',skiprows=1).set_index('No')
DF_CUin_Value = pd.read_excel(ExcelFile, sheet_name='CUin_Value',skiprows=1).set_index('No')
DF_CUout = pd.read_excel(ExcelFile, sheet_name='CUout',skiprows=1).set_index('No')
DF_CUout_Value = pd.read_excel(ExcelFile, sheet_name='CUout_Value',skiprows=1).set_index('No')
DF_CF = pd.read_excel(ExcelFile, sheet_name='CF',skiprows=1).set_index('No')
DF_CF_Value = pd.read_excel(ExcelFile, sheet_name='CF_Value',skiprows=1).set_index('No')
#print(DF_RP.iloc[0,1:len(Factor_RP)])
#print(x)
#print(y)
```

```python
def draw_Plot(DF, Title, FigName):
    %matplotlib inline
    # set plot area
    fig_high = 10
    fig_width = 15
    fig, (ax) = plt.subplots(1,1,figsize=(fig_width,fig_high),
                                   gridspec_kw=dict(width_ratios=[1],height_ratios=[1],
                                                    wspace=0.30, hspace=0.14)) 

    plt.rcParams['font.family'] = 'Arial'
    
    #set initial
    fontsize = 18
    labelsize = 15
    
    Factor=DF.columns.drop('Factor') #X
    x = np.arange(len(Factor))
    
    ## write x=0 line
    ax.axhline(linestyle='--', linewidth=0.5, color='black')
    
    
    for i in range(len(DF.index)):
        y = DF.iloc[i, 1:len(x)+1]
        if i < 29:
            ax.plot(x,y,marker="o", linewidth=0 , ms=10, mfc="#d4e3f9", mec="black", mew=0.5)
    
        elif i == 29:
            ax.plot(x,y,marker="o", linewidth=0 , ms=10, mfc="#d4e3f9", mec="black", mew=0.5,
                    label = 'Average of 1 pair out of 30 pairs')
        
        elif i == 30:
            ax.plot(x,y,marker="o", linewidth=0 , ms=15,mfc="#2a5ea8", mec="black", mew=0.75,
                    label = 'Average of 30 pairs')
        
        else: #i == 31:
            ax.plot(x,y,marker="o", linewidth=0 , ms=7.5, mfc="#d3354c", mec="#d3354c", 
                    label = 'Average of 100 independent pairs')


    ## set figure elements
    x_label = []
    for i in range(len(x)):
        x_label.append(Factor[i])
        
    ax.tick_params(labelsize=labelsize)
    if len(x)<15:
        ax.set_xticklabels(x_label, fontsize=labelsize, ha='center', rotation=0)
    else:
        ax.set_xticklabels(x_label, fontsize=labelsize, ha='center', rotation=-45)
    ax.set_xticks(x)
    ax.yaxis.set_major_formatter(mpl.ticker.PercentFormatter(1))
    ax.set_ylabel('Relative difference from average of 100 indipendent pairs (%)',size=fontsize)
    
    ax.legend(fontsize=labelsize)
    
    ax.set_title(Title, size=fontsize + 2, loc='left')
    
    # Save figure
    fig.savefig(FigName, format="svg", bbox_inches="tight", facecolor='none')
```
