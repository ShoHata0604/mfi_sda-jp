```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import math
from matplotlib.ticker import *
def draw_piechart(x,y,c,label):
    """ドーナツグラフを表示する関数
    引数：
    x：x軸の値（ラベル）
    y：y軸の値（表示する数字）
    c：指定色
    label：ラベル名
    """
    #グラフの装飾
    plt.title(label,fontsize = 25,weight="bold",x=0.5,y=0.47)#中心に系列名を表示
    plt.rcParams['font.size'] = 20
    edgecolor = 'black'
    C_none = ['#4169e100', '#4169e100', '#4169e100']
    Linewidth = 1.5
    Percentage=[]
    for i in range(len(y)):
        Percentage.append('{:.0%}'.format(y.iloc[i]))
    #外側のグラフの描画
    patches, texts = plt.pie(y,counterclock=False,startangle=90,colors=c,
                         labeldistance=1.15,wedgeprops={'linewidth': Linewidth, 'edgecolor':edgecolor},
                         textprops={'color': "black", 'weight': "bold"}) 
    for t in texts:
        t.set_horizontalalignment('center') #ラベルの表示場所の指定　ラベルを表示しない場合は無関係
    #外側のグラフの枠線を描写（内円の割合の枠）
    #patches, texts = plt.pie(y2,labels=x2,counterclock=False,startangle=90,colors=C_none,
    #                     wedgeprops={'linewidth': Linewidth, 'edgecolor':edgecolor},radius=1,labeldistance=1.5,#ラベルを白くして見えなくしている
    #                     textprops={'color': 'White', 'weight': "bold"})
    
    #内側のグラフの描画（内円の大きさはradiusで指定、ラベルの位置はlabeldistanceで指定）
    #patches, texts = plt.pie(y2,labels=x2,counterclock=False,startangle=90,colors=c2,
    #                     wedgeprops={'linewidth': Linewidth, 'edgecolor':edgecolor},radius=0.75,labeldistance=1.5,#ラベルを白くして見えなくしている
    #                     textprops={'color': 'White', 'weight': "bold"})
    #for t in texts:
    #    t.set_horizontalalignment('center')
        
    #中心に白円を挿入
    center_circle = plt.Circle((0, 0), 0.3, color=edgecolor, fc='white', linewidth=Linewidth) #２つ目の引数が円の大きさ
    fig = plt.gcf()
    fig.gca().add_artist(center_circle)

## Set DataFrame
ExcelFile = '/Users/hata_s/Dropbox/01_Research/02_MF_TimeSeries/03_Matlab/02_Result/R4_Data.xlsx'
## Resource productivity
DF_RP = pd.read_excel(ExcelFile, sheet_name='RP',skiprows=1).set_index('No')
DF_RP = DF_RP.reindex(index=[2,3,1,4,5])
DF_RP.loc[6] = DF_RP.iloc[2,:]+DF_RP.iloc[3,:]+DF_RP.iloc[4,:]
DF_RP.iloc[5,0] = 'MICI+MDCD+Non'
DF_RP = DF_RP.reindex(index=[2,3,6,1,4,5])
DF_RP = DF_RP.iloc[0:3,:]

DF_CUin = pd.read_excel(ExcelFile, sheet_name='CUin',skiprows=1).set_index('No')
DF_CUin = DF_CUin.reindex(index=[2,3,1,4,5])
DF_CUin.loc[6] = DF_CUin.iloc[2,:]+DF_CUin.iloc[3,:]+DF_CUin.iloc[4,:]
DF_CUin.iloc[5,0] = 'MICI+MDCD+Non'
DF_CUin = DF_CUin.reindex(index=[2,3,6,1,4,5])
DF_CUin = DF_CUin.iloc[0:3,:]

DF_CUout = pd.read_excel(ExcelFile, sheet_name='CUout',skiprows=1).set_index('No')
DF_CUout = DF_CUout.reindex(index=[2,3,1,4,5])
DF_CUout.loc[6] = DF_CUout.iloc[2,:]+DF_CUout.iloc[3,:]+DF_CUout.iloc[4,:]
DF_CUout.iloc[5,0] = 'MICI+MDCD+Non'
DF_CUout = DF_CUout.reindex(index=[2,3,6,1,4,5])
DF_CUout = DF_CUout.iloc[0:3,:]

DF_FD = pd.read_excel(ExcelFile, sheet_name='FD',skiprows=1).set_index('No')
DF_FD = DF_FD.reindex(index=[2,3,1,4,5])
DF_FD.loc[6] = DF_FD.iloc[2,:]+DF_FD.iloc[3,:]+DF_FD.iloc[4,:]
DF_FD.iloc[5,0] = 'MICI+MDCD+Non'
DF_FD = DF_FD.reindex(index=[2,3,6,1,4,5])
DF_FD = DF_FD.iloc[0:3,:]

print(DF_RP) 
print(DF_CUin)
print(DF_CUout)
print(DF_FD)
```

```python
%matplotlib inline
from matplotlib.colors import ListedColormap, BoundaryNorm

plt.rcParams['font.family'] = 'Arial'

# set plot area
fig_high = 30
fig_width = 25
wspace = 0.01

fig = plt.figure(figsize=(fig_width, fig_high))
#fig1,(ax0, ax1, ax2, ax3) = plt.subplots(1,4,figsize=(fig_width,fig_high),
#                                    gridspec_kw={"width_ratios":[1,1,1,1]})#各ax、caxの横の長さを比率で設定
plt.subplots_adjust( wspace=wspace, hspace=0) # 間隔指定

# set initials
fontsize = 22
labelsize = 20

Color_MGCB = '#5e95e5'
Color_MBCG = '#fcda4f'
Color_Other = 'lightgrey'
Colors = [Color_MGCB,Color_MBCG,Color_Other]

# draw piechart
# RP
Label='RP'
plt.subplot(4,4,1)
draw_piechart(DF_RP.INDEX,DF_RP.D1,Colors,Label)
plt.subplot(4,4,5)
draw_piechart(DF_RP.INDEX,DF_RP.D2,Colors,Label)
plt.subplot(4,4,9)
draw_piechart(DF_RP.INDEX,DF_RP.D3,Colors,Label)
plt.subplot(4,4,13)
draw_piechart(DF_RP.INDEX,DF_RP.y,Colors,Label)
# FD
Label='FD'
plt.subplot(4,4,2)
draw_piechart(DF_FD.INDEX,DF_FD.D1,Colors,Label)
plt.subplot(4,4,6)
draw_piechart(DF_FD.INDEX,DF_FD.D2,Colors,Label)
plt.subplot(4,4,10)
draw_piechart(DF_FD.INDEX,DF_FD.D3,Colors,Label)
plt.subplot(4,4,14)
draw_piechart(DF_FD.INDEX,DF_FD.y,Colors,Label)
# CUin
Label='CU$_{\mathrm{in}}$'
plt.subplot(4,4,3)
draw_piechart(DF_CUin.INDEX,DF_CUin.D1,Colors,Label)
plt.subplot(4,4,7)
draw_piechart(DF_CUin.INDEX,DF_CUin.D2,Colors,Label)
plt.subplot(4,4,11)
draw_piechart(DF_CUin.INDEX,DF_CUin.D3,Colors,Label)
plt.subplot(4,4,15)
draw_piechart(DF_CUin.INDEX,DF_CUin.y,Colors,Label)
# CUout
Label='CU$_{\mathrm{out}}$'
plt.subplot(4,4,4)
draw_piechart(DF_CUout.INDEX,DF_CUout.D1,Colors,Label)
plt.legend(DF_RP.INDEX, fontsize=18,bbox_to_anchor=(0.9, 0.9),facecolor='none',frameon=False,borderpad=0.8,labelspacing=0.7) 
plt.subplot(4,4,8)
draw_piechart(DF_CUout.INDEX,DF_CUout.D2,Colors,Label)
plt.subplot(4,4,12)
draw_piechart(DF_CUout.INDEX,DF_CUout.D3,Colors,Label)
plt.subplot(4,4,16)
draw_piechart(DF_CUout.INDEX,DF_CUout.y,Colors,Label)

# Save figure
fig.savefig("/Users/hata_s/Dropbox/02_Python/03_MFI_SDA/00_Figfile/MFI_SDA_R4.svg",
             format="svg", bbox_inches="tight", facecolor='none')
```
